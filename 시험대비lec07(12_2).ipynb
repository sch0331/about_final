```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'Date': ['2024-10-01','2024-10-01','2024-10-02','2024-10-02','2024-10-03','2024-10-03'],
    'Product': ['A','B','A','B','A','B'],
    'Sales': [150,200,180,220,170,210]
    }
```

```python
#Anaylzing data of Product in A, B
df = pd.DataFrame(data)
df_A = df[df['Product'] == 'A']   #df_A를 새롭게 define/ A인거만 dataframe!
df_B = df[df['Product'] == 'B']
```

```python
#Anaylzing sales data
print("Total Sales for Product A: ", sum(df_A['Sales']))
print("Total Sales for Product B: ", sum(df_B['Sales']))

average_A = np.mean(df_A['Sales']) #np.mean은 df_A['Sales']의 평균을 구하는 식
average_B = np.mean(df_B['Sales'])
print(f"Average Sales for Product A: {average_A:.2f}")
print(f"Average Sales for Product B: {average_B:.2f}")
```

```python
#Visualization
plt.figure(figsize=(10,6))
plt.plot(df_A['Date'], df_A['Sales'], marker ='o', label = 'A')
plt.plot(df_B['Date'], df_B['Sales'], marker ='o', label = 'B')
plt.title('Daily Sales Trend by Product')
plt.legend(title = 'Product')
plt.show()
```

```python
total_sales = df.groupby('Product')['Sales'].sum() #Product에서 동일한 이름을 가지는 것을 묶어서 'Sales'의 총합을 구함.

plt.figure(figsize=(10,6))
plt.bar(total_sales.index, total_sales.values)
plt.title('Total Sales by Product')
```
