```python
import json
import pandas as pd
```

```python
with open('input_large_inventory.json', 'r') as f:
  data = json.load(f)

df = pd.DataFrame(data)
```

```python
df['total_value'] = df['quantity'] * df['price']
```

```python
price_min = df['price'].min()
price_max = df['price'].max()
df['price'] = (df['price'] - price_min)/(price_max - price_min)
```

```python
df["demand_label"] = df["demand_label"].map({"high": 1, "low": 0})
```

```python
df_filtered = df[df["quantity"] >= 10]
```

```python
mean_high = df_filtered[df_filtered["demand_label"] == 1]["total_value"].mean()
print(mean_high)

dist = df_filtered["demand_label"].value_counts()
print(dist)
```
